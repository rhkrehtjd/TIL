# 음이항 분포

- review : 기하분포란 성공확률을 p라고 했을 때 x번째 시행에서 첫 번째 성공이 나올 p(x), 즉 x-1번 시행은 전부 실패.
- 음이항 분포는 이 기하분포의 확장 버전이다. 
- 음이항 분포는 총 다섯가지의 정의가 있는데 그 중 네 번째가 기하분포의 확장이다.
- 네 번째 정의 : 성공확률이 p이며, x번째 시행에서 k번 성공이 나올 p(x), 여기서 k는 변수가 아님. 결정되는 값임. 예를 들어 k가 3일 때, x번의 시행에서 마지막은 성공으로 끝나야하고 마지막 시행 제외하고 앞의 x-1번의 시행 중 성공이 2번 더 나와야하는 것임.
- 이 분포를 함수로 작성해보면,

- 
- 즉 기하분포라는 것은 k가 1일 때의 분포가 되는 것임, 기하분포는 음이항 분포에 포함될 것임

---

- 또 다른 정의
- x 번째 시행이라는 것은 전체 시행 횟수인데 이때 이 전체 시행횟수를 n 성공 횟수를 k라 해보자. 그리고 실패 횟수는 r.
- n=k+r 

(1) r이 정해지고 k가 변수인 경우
- 가장 많이 사용됨
- r번의 실패가 나오기까지 발생한 성공이 k번일 확률
- 예를 들어 r이 3일 때 k = 0,1,2,...가 될 것,
- r이 3이고 k가 1이라면 마지막은 실패여야 하고 앞에 성공이 한 번 나올 것(실실성실)

(2) r이 정해지고 n이 변수인 경우

(3) k가 정해지고 r이 변수인 경우

(4) k가 정해지고 n이 변수인 경우 
- (위에서 설명한 경우임)

(5) n이 정해지고 k 또는 r이 변수인 경우
- 이항분포와 같다.

### 음이항 분포의 예시
 review : 우리는 지난시간에 음이항 분포에는 5가지가 있다고 배웠다. 이 5가지의 음이항 분포는 n=k+r를 어떤 값을 독립, 종속, 상수로 놓을지에 따라 나눈다고 했다. 이 다섯가지의 음이항 분포의 정의 중에서 첫 번째 정의를 많이 사용한다. 첫 번째 경우의 예를 살펴보자
 
 - 어떤 농구선수가 있고 이 농구선수의 자유투 성공률이 30%일 때 이때 이 농구선수가 세번의 실패가 나오기까지(r이 정해졌음) 발생한 성공이 x번일 확률의 분포가 음이항 분포(p(x))이다.
 - x=0일 땐 발생한 성공이 0번 즉 실패만 3번 따라서 p(0)=(0.7)^3
 - x=1일 땐 발생한 성공이 1번 즉 마지막은 실패로 꼭 끝나야 함 (실패2성공1)실패, 따라서 3combination1*0.7^2*0.3*0.7
 - x=2일 땐 마찬가지로 실패로 끝나며, 앞의 네개의 자리에 실패 두개 성공 두개가 나와야 함. 위의 경우처럼 조합을 이용하여 표현할 수 있다.

### 음이항 분포 유도

- 어떤 사건의 발생 확률이 p일 때 r번의 실패가 나오기까지 발생한 성공이 k번인 확률 분포
- f(k; r,p) 이때 세미콜론 앞에 있는 k는 독립 변수이고 세미 콜론 뒤 문자들은 상수이다. 즉 실패횟수와 성공확률이 정해져 있는 상황에서 독립변수인 k가 변하면서 나타나는 분포이다. 
- f(x;r,p)=p(x)
- r번째 실패로 끝이 났고, 그 전엔 성공은 x번 실패는 r-1번, 즉, 마지막 시행 전엔 x+r-1번의 시행이 일어났다. 이때 x번의 성공을 배열하고 총 실패는 마지막까지 포함해서 r번했으니 확률분포함수로 나타내면
