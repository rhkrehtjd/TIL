# 이항 분포

- 이항분포의 설명은 베르누이 시행으로 시작한다. 베르누이 시행은 시행의 결과가 오직 2개뿐인 시행을 말한다. 성공 혹은 실패, 사건의 발생 혹은 발생하지 않음, 동전의 앞면 혹은 뒷면 이렇게 사건의 결과가 오직 둘로만 나타는 시행을 베르누이 시행이라고 한다. 이러한 베르누이 시행을 여러번 한 것이며 각각의 시행은 독립이다. 원래 베르누이 시행엔 독립이 포함되어 있기 때문에 독립을 언급해줄 필요는 없다. 
- 이때 베르누이 시행을 n번 수행해주고 이 사건이 발생한 횟수를 확률 변수로 하는 분포가 이항 분포이다.
- 예를 들어보자, 어떤 농구선수가 있는데 이 농구선수의 자유투 성공률이 0.8일 때 이 농구선수가 자유투를 10번 던질 때 실패할 수도 성공할 수도 있을텐데 이때 이 농구선수가 자유투를 성공하는 횟수를 확률변수 X라고 놓고 X라는 횟수만큼 성공할 확률을 P(X)라 놓자. 
- P(0)를 먼저 구해보자 : 10번 다 실패한 확률 = $0.2^10$
- P(1)를 구해보자 : 1번만 성공했다. = $0.8 \times 0.2^9$ 그런데 이렇게 한 번만 성공하고 아홉번만 실패하는 경우는 총 열가지의 경우가 있으므로 $0.8 \times 0.2^9$ 여기에 10을 곱해줘야 함
- P(2)를 구해보자 : $0.8^2 \times 0.2^8$ 열번 중에 2번 성공했으므로 조합의 표현을 이용해보면 10combination2를 곱해주어야 한다.

- 이항 분포의 그래프
      n이 일정할 때 확률이 그래프에 미치는 영향을 볼록한 부분의 위치임을 알 수 있고 n이 커지면 커질 수록 종모양이 되는 걸 볼 수 있는데, 이항 분포에서 n이 커지면 커질수록 정규분포에 근사해감을 알 수 있다. 그런데 정규 분포로 근사할 수 있다는 건 p가 너무 크거나 너무 작지 않을 때 가능하다. 이때를 skewness 즉 왜도가 크다. 즉, 비대칭도가 커져서 정규 분포로 근사하긴 어렵다. 대신 n이 충분히 클 때 p가 0에 가까우면 푸아송 분포에 근사할 수 있다. 나중에 푸아송 분포 때 다시 언급하겠다.

```python
import numpy as np
from math import factorial
def bin_dist(k, n, p):
    nck = factorial(n) / (factorial(k) * factorial(n - k))
    pd = nck * p**k * (1-p)**(n-k)
    return pd

x = np.arange(11)
pd1 = np.array([bin_dist(k, 10, 0.1) for k in range(11)])
ax1.scatter(x,pd1)
ax1.set_title('n=10,p=0.1')
x = np.arange(11)
pd1 = np.array([bin_dist(k, 10, 0.3) for k in range(11)])
ax2.scatter(x,pd1)
ax2.set_title('n=10,p=0.3')
x = np.arange(11)
pd1 = np.array([bin_dist(k, 10, 0.5) for k in range(11)])
ax3.scatter(x,pd1)
ax3.set_title('n=10,p=0.5')
x = np.arange(11)
pd1 = np.array([bin_dist(k, 10, 0.7) for k in range(11)])
ax4.scatter(x,pd1)
ax4.set_title('n=10,p=0.7')
x = np.arange(11)
pd1 = np.array([bin_dist(k, 10, 0.9) for k in range(11)])
ax5.scatter(x,pd1)
ax5.set_title('n=10,p=0.7')
x = np.arange(31)
pd1 = np.array([bin_dist(k, 30, 0.1) for k in range(31)])
ax6.scatter(x,pd1)
ax6.set_title('n=30,p=0.1')
x = np.arange(31)
pd1 = np.array([bin_dist(k, 30, 0.3) for k in range(31)])
ax7.scatter(x,pd1)
ax7.set_title('n=30,p=0.3')
x = np.arange(31)
pd1 = np.array([bin_dist(k, 30, 0.5) for k in range(31)])
ax8.scatter(x,pd1)
ax8.set_title('n=30,p=0.5')
x = np.arange(31)
pd1 = np.array([bin_dist(k, 30, 0.7) for k in range(31)])
ax9.scatter(x,pd1)
ax9.set_title('n=30,p=0.7')
x = np.arange(31)
pd1 = np.array([bin_dist(k, 30, 0.9) for k in range(31)])
ax10.scatter(x,pd1)
ax10.set_title('n=30,p=0.9')
x = np.arange(101)
pd1 = np.array([bin_dist(k, 100, 0.1) for k in range(101)])
ax11.scatter(x,pd1)
ax11.set_title('n=100,p=0.1')
x = np.arange(101)
pd1 = np.array([bin_dist(k, 100, 0.3) for k in range(101)])
ax12.scatter(x,pd1)
ax12.set_title('n=100,p=0.3')
x = np.arange(101)
pd1 = np.array([bin_dist(k, 100, 0.5) for k in range(101)])
ax13.scatter(x,pd1)
ax13.set_title('n=100,p=0.5')
x = np.arange(101)
pd1 = np.array([bin_dist(k, 100, 0.7) for k in range(101)])
ax14.scatter(x,pd1)
ax14.set_title('n=100,p=0.7')
x = np.arange(101)
pd1 = np.array([bin_dist(k, 100, 0.9) for k in range(101)])
ax15.scatter(x,pd1)
ax15.set_title('n=100,p=0.9')
x = np.arange(1001)
pd1 = np.array([bin_dist(k, 1000, 0.1) for k in range(1001)])
ax16.scatter(x,pd1)
ax16.set_title('n=1000,p=0.1')
x = np.arange(1001)
pd1 = np.array([bin_dist(k, 1000, 0.3) for k in range(1001)])
ax17.scatter(x,pd1)
ax17.set_title('n=1000,p=0.3')
x = np.arange(1001)
pd1 = np.array([bin_dist(k, 1000, 0.5) for k in range(1001)])
ax18.scatter(x,pd1)
ax18.set_title('n=1000,p=0.5')
x = np.arange(1001)
pd1 = np.array([bin_dist(k, 1000, 0.7) for k in range(1001)])
ax19.scatter(x,pd1)
ax19.set_title('n=1000,p=0.7')
x = np.arange(1001)
pd1 = np.array([bin_dist(k, 1000, 0.9) for k in range(1001)])
ax20.scatter(x,pd1)
ax20.set_title('n=1000,p=0.9')
fig
```
