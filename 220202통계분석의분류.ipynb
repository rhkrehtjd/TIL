# 통계분석의 분류

- 통계분석 방법은 선택할 때 고려해야 할 사항들이 있다. 정리해보면 아래와 같다.
  - 분석의 목적
  - 집단의 수
  - 독립변수의 종류
  - 종속변수의 종류
  - 정규성 여부
  - 등분산 여부 등이 있다.
- 이와 같은 사항들이 통계분석을 분류하는 데 있어서 분기의 기준이 될 것이다. 어느 사항으로 먼저 분기하냐에 따라서 정말 수 많은 종류의 분류차트가 나올 수 있다.
- 모든 조합을 고려하여 수 많은 차트를 그려내면 오히려 헷갈릴 수 있기 때문에 나름의 기준을 정하고 분류해보자
- 분류 대상이 될 통계 분석 방법들은 아래와 같다.
  - 정규성 검정
  - F 검정
  - Levene's 검정
  - 독립표본t검정(등분산)
  - 독립표본t검정(이분산)
  - 대응표본t검정
  - 월콕슨부호순위검정
  - 일원분산분석
  - 이원분산분석
  - 다원분산분석
  - 다변량분산분석
  - 반복측정분산분석
  - Kruskal wallis test
  - Friedman test
  - 비율검정
  - 카이제곱검정
  - 피션의 정확검정
  - 선형 대 선형 결합(카이제곱 트렌드 검정)
  - 피어슨 상관 분석
  - 스피어만 상관분석
  - 캔달의 타우 상관분석
  - 단순 선형 단변량 회귀분석
  - 단순 비선형 단변량 회귀분석
  - 단순 선형 다변량 회귀분석
  - 단순 비선형 다변량 회귀분석
  - 단순 선형 다변량 회귀분석
  - 단순 비선형 다변량 회귀분석
  - ICC
  - 코헨의 카파
  - 검정력 분석
  - ADF 검정

- 두 집단의 평균을 비교할 때 사용하는 검정방법들은 아래와 같다. 
  - z 검정
  - 독립표본 t검정
  - 대응표본 t검정
  - 윌콕슨 순위합 검정
  - 윌콕슨 부호순위검정
- 어떤 조건에서 어떤 검정 방법을 선택해야하는지 알아보자. 
- 두 집단의 평균을 비교하는 검정 방법을 선택할 때 고려해야할 사항들은 아래와 같다.
  - 모분산을 아는가?
    - 모분산을 알아야 z검정 사용이 가능하다. 모분산은 보통 모르기 때문에 주로 t검정이 사용된다. 만약 표본의 크기가 아주아주 크다면, 표본분산을 모분산 대신 사용할 수 있다만, t검정이 등장하기 전에 그랬던 것이기에 그냥 t검정을 사용하면 된다. 표본 크기가 커지면 z검정을 쓰나 t검정을 쓰나 결과는 동일하다. 
  - 정규성을 가정할 수 있는가?
    - 정규성의 가정은 일반적으로 표본의 크기가 30이상이면 가능하다. 표본의 크기가 10~30이면 정규성 검증을 통해 정규성 여부를 판단한다. 표본의 크기가 10보다 작은 경우 비모수 검정을 수행한다.
    - 두 집단 평균 비교의 비모수 검정은 주로 LF test(Lilliefors test)나 SW test (Shapiro-wilk-test)가 사용된다.
  - 두 표본이 서로 독립인가 대응인가?
    - 대응표본은 '투약 전 후','처리 전 후'와 같이 표본이 대응되는 경우를 의미한다.
  - 등분산 가정이 가능한가?
    - 독립표본 t검정은 등분산성 여부에 따라 둘로 나뉜다. 등분산가정 t검정과 이분산가정 t검정이다. 이분산 가정 t검정은 Welch's t-test라고도 부른다.
