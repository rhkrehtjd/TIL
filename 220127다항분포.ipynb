# 다항 분포

- `이`항분포는 이름에서도 알 수 있듯 시행의 결과가 두가지인 분포를 의미한다. 사건이 발생했거나, 사건이 발생하지 않았거나 이런식으로 말이다. 사건 발생확률을 p라고 한다면 발생하지 않을 확률은 1-p일 것이다.
- 이항분포의 확률을 구해보자. 시행을 10번 했을 때 6의 눈이 3번 나올 확률을 구해보자. P(3) = 10C3*(1/6)^3(5/6)^7이다.
- 이항분포와 다르게 다항분포는 시행의 결과가 셋 이상이다.
  - 주사위 예시를 들어보자
     - 주사위 한 개를 던졌을 때 시행의 결과가 1의 눈이 발생하는 경우 2,3,4의 눈이 발생하는 경우, 5의 눈이 발생하는 경우, 6의 눈이 발생하는 경우
     - 이 예시에서 다항분포의 확률을 구해보자
     - 시행을 10번 했을 때, 1의 눈이 3번 2,3,4의 눈이 5번, 5의 눈이 1번, 6의 눈이 1번.
     - 이것을 확률로 나타내면 P(3,5,1,1)로 나타낼 수 있다. 
     - (계수)(1/6)^3(1/2)^5(1/6)(1/6)
     - 계수 = 10!/3!5!1!1!
     - 즉, 계수는 같은 것이 있는 순열을 이용하여 구할 수 있다.

- 다항분포를 유도해보자
  - 어떤 시행을 했을 때 시행의 결과로 일어날 수 있는 사건이 K개라고 해보자. 그렇다면 사건1부터 사건K까지 있을 것이다. 이 한 번의 시행에서는 하나의 사건만 발생할 수 있다. 이때 사건1이 발생할 확률은 P(1), 사건2가 발생할 확률은 P(2),, 이런식으로 나타낼 수 있으며 이 모든 확률을 더했을 때 1이 나올 것이다. 우리가 이 시행을 n번 했을 때 사건1이 발생한 횟수가 x_1회 사건2가 발생한 횟수가 x_2회,, 이렇게 표현할 수 있을 것이며 이 횟수를 모두 더하면 n이 나올 것이다.
      - 확률을 한 번 구해보자
      - 시행을 n번 했을 때, 사건1이 x_1번, 사건2가 x_2번,,,발생할 확률이다.
      - P(x_1,x_2,...,x_k)로 나타낼 수 있겠다. 
      - (계수)P(1)^x_1*P(2)^x_2*...*P(k)^x_k
      - 계수는 전에 설명했듯 같은 것이 있는 순열로 구할 수 있겠다.
      - 
      - 다항 분포의 확률 분포식

- 다항분포의 예시를 알아보자
  - 상자가 한 개 있다. 상자 안에는 공이 100개 들어있고 빨강공이 20개 파란공이 30개 노란공이 50개 들어있다. 이제 상자에서 공을 10번 추출할 건데 복원추출로 한다. 이렇게 10번 공을 뽑았을 때 10번 뽑힌 공 중에서 빨간공이 5번 파란공이 2번 노란공이 3번 뽑혔을 확률을 구해보자.
      - 먼저 공을 한 번 뽑는 상황을 생각해보자
      - 공을 한 번 뽑았는데 그 공이 빨간공일 확률은 얼마일까. 20/100=0.2
      - 공을 한 번 뽑았는데 그 공이 파란공일 확률은 얼마일까. 0.3
      - 공을 한 번 뽑았는데 그 공이 노란공일 확률은 얼마일까. 0.5
      - 이번에는 다항분포를 적용해보자
      - 공을 n번 뽑았을 때 빨간공이 x번, 파란공이 y번, 노란공이 z번 뽑힐 확률을 한 번 구해보자.
      - 
      - 이 확률이 다항분포이다.

- 다항분포의 평균과 분산에 대해 알아보자
  - 다항분포의 기댓값을 각 사건별로 기댓값을 구하거나 사건들의 합집합의 기댓값을 구할 수 있다. 
  - 한 가지 다항분포를 예를 들어보자
      - 어떤 시행을 했을 때 이 시행의 결과가 A라는 사건 B라는 사건 C라는 사건이 되는 시행이다. 그리고 각 사건이 발생할 확률은 P(A),P(B),P(C)라고 놓자. 
      - 그러면 P(A)+P(B)+P(C) = 1 이 된다. 
      - 이 시행을 n번 수행 했을 때 각 사건 A,B,C가 발생한 횟수를 x,y,z로 놓자.
      - x+y+z=n 이 될 것이다.
      - 이번에는 n번의 시행에서 사건 A가 x번, 사건 B가 y번, 사건 C가 z번 발생할 확률을 구해보자
      - 
      - 지금부터 이 다항분포를 이용해서 기댓값을 구해보자
           - 먼저 사건 A의 기댓값을 구해보자, 사건 A의 기댓값을 E(X)로 나타낸다. 
           - 우리가 위에서 예로 든 시행은 시행의 결과 A,B,C인데 이 시행을 A의 관점에서 바라보면 이 시행의 결과를 A가 발생하거나 A가 발생하지 않거나 이렇게 두가지로만 나눌 수 있고 이렇게 두가지로 나눠지는 확률 분포는 이항 분포였다. 그래서 우리는 위에 있는 다항분포를 A에 대한 이항분포로 이해할 수 있다. 그러면 이항분포의 평균을 구해주면 우리가 원하던 사건 A의 평균이 될 것이다.
           - 따라서 사건 A의 평균은 n*P(A) 분산은 n*P(A)*(1-P(A))로 구할 수 있다. 
           - 사건 B와 사건 C도 마찬가지로 구할 수 있다.
           - E(Y) = n*P(B), E(Z) = n*P(C) 이렇게 구할 수 있겠다. 
       - 이번에는 어떤 두 사건의 교집합을 생각해보자.
           - 예를 들면, A와 B의 교집합인데, A와 B의 교집합은 X와 Y의 교집합의 기댓값이고 이 교집합의 기댓값은 항상 0이다. 왜? A와 B가 서로 배반사건이기 때문이다. A가 발생하면 B가 절대 발생할 수 없고, B가 발생하면 A가 절대 발생할 수 없기 때문이다. 따라서 교집합은 절대 발생하지 않는다.  
           - 우리가 실행의 결과를 몇 개의 사건으로 나눴든 상관없이 이 시행의 결과로 발생하는 사건들의 교집합의 기댓값을 구하면 그 결과는 항상 0이 된다. 
       - 이번에는 합집합을 생각해보자
            - A와 B의 합집합의 기댓값,A 또는 B가 발생하는 것에 대한 기댓값.
            - 
            - 위에서 이때 교집합이 발생할 확률은 0이라고 했으므로 이 둘의 합집합을 구할 때 따로 차감해줄 필요는 없다. 
            - 우리는 이 시행의 결과를 A또는 B가 발생하거나 A또는 B가 발생하지 않거나. 이렇게 이항분포로 생가갈 수 있다.
            - 그래서 이 기댓값은 
            - 사건의 개수와 무관하게 A와 B와 C의 합집합의 기댓값을 구할 때도 동일하게 적용할 수 있다.

- 다항분포의 독립 변수의 개수에 대해서 알아보자
  - 어떤 시행의 결과가 A,B,C인 다항분포에서 독립변수의 개수는 몇개일까
  - 3개가 아닌 2개이다. $WHY$?
      - 먼저 시행의 결과가 2가지인 경우부터 생각해보자. 사건 A, 사건 B.
      - p(A)+p(B)=1
      - 이런 시행을 n번 했을 때 사건 A가 x번 발생하고 사건 B가 y번 발생할 확률을 한 번 구해보자
      - 
      - 얼핏 봤을 땐 변수가 2개 들어있어서 x,y 둘다 독립변수처럼 보이지만 사실 둘 중 하나는 종속변수이다. 왜냐면 x+y=n이라는 등식이 성립하기 때문에 x가 결정이 되면 y는 저절로 결정이 되고 반대로 y가 결정이 되면 x가 저절로 결정이 된다. 
      - 따라서 우리는 이 식을 y=n-x 이렇게 바꿀 수 있다. 
      - 이렇게 바뀐 식을 이용하여 다시 써보자.
      - 
      - 이항분포의 분포함수식이 됐다.이항분포도 어쨌든 시행의 결과가 2이기 때문에.
      - 이 이항분포에서 이항이라는 말은 시행의 결과가 2가지를 뜻이다. 우리는 이 이항분포를 시행의 결과가 2가지인 다항분포라고 말할 수도 있겠다. 
      - 이번에는 시행의 결과가 3가지인 다항분포를 생각해보자
      - 시행의 결과가 A,B,C이며, 각각의 사건이 발생할 확률을 P(A),P(B),P(C)로 놓고 다 더하면 1이 될 것이다.
      - 마찬가지로 n번의 시행을 했을 때 A가 x번, B가 y번, C가 z번 발생했다고 하자
      - 이때 확률분포 함수식
      - 
      - z = n - x - y 로 나타낼 수 있다. z를 x와 y의 종속변수로 보는 것이다.
      - 
      - 독립변수가 2개인 `삼항분포`라고 부른다.
      - 시행의 결과가 m개의 사건으로 나누어지는 다항분포에서 독립변수의 개수는 m-1개가 된다.

- 다항분포의 그래프에 대해 알아보자
  - 시행의 결과가 2가지인 그래프
  - 
  - 우리가 이 그래프에서 알 수 있는 것은 사건 발생 확률이 커질 수록 봉우리 부분이 오른쪽으로 이동한다는 것이고, 시행횟수가 커질수록 봉우리 부분이 오른쪽으로 이동하는데, 시행횟수가 커지면 그래프의 모양이 좌우 대칭인 종모양이 된다는 것을 알 수 있다. 
  - 실제로 이항분포는 n이 무한대로 가면 정규분포에 가까워져 간다.
  - 이번에는 독립변수를 2개로 늘린, 즉 시행의 결과가 3가지인 그래프에 대해 알아보자
  - 
  - 시행 횟수를 늘려보자
  - 
  - 종모양의 그래프에 근접해가는 것 같다. 
  - 이항분포에서 n이 커질수록 정규분포에 가까워져 간다고 했는데 차원을 하나 늘려도 동일한 원리가 성립함을 알 수 있다.
  - 마지막으로 사건 B의 확률을 0.5로 늘려보자
  - 그러면 기댓값이 25로 증가하게 되고 그래프가 어떻게 변할지 미리 예측해보자면, 봉우리의 위치가 y가 25로 이동하게 될 것이다.
  -
