# 기하 분포

- 기하분포도 이항분포와 마찬가지로 기본 단위가 베르누이 시행(시행의 결과가 오직 두개)이다. 
- 베르누이 시행의 결과가 계속 실패가 나오다가 성공이 나왔을 때 성공이 나오기까지 총 시행한 횟수를 확률 변수 X로 하는 분포가 기하 분포이다.
- 예를 들면 확률 변수 X=4인 경우는 실패-실패-실패-성공 이런 경우이다.
- 그런데 이런 기하분포에는 정의가 두가지 있다.
      
      첫 번째 정의는 위에서 설명한 정의이고
      
      두번 째 정의는 처음 성공이 나오기까지 실패한 횟수를 확률변수 X로 하는 확률 분포이다.

- 두 번째 정의에 따르면, 확률변수 X=4인 경우는 실패-실패-실패-실패-성공 이런 경우일 것이다. 이 경우를 첫 번째 정의를 적용하면 X=5가 될 것이다.
- 지금부터는 첫 번째 정의를 통해 설명하겠다.

- 남자와 여자가 연애 후 결혼할 확률이 5%라고 할 때 X번 째 사귄 이성과 결혼하게 될 확률을 P(X)라 한다면 이 때 P(X)가 기하분포가 될 것이다. P(1)은 첫 번째 여성과 결혼할 확률이므로 P(1)=0.05이고 P(2)는 두 번째 여성과 결혼할 확률이므로 P(2)=0.95 X 0.05 이다. 이런 분포가 기하분포이다.

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x,p):
    return ((1-p)**(x-1))*p

plt.figure(figsize=(10,7))
x=np.arange(1,11)
narr=np.array([f(k, 0.1) for k in np.arange(1,11)])
plt.plot(x,narr,'.')
x=np.arange(1,11)
narr=np.array([f(k, 0.3) for k in np.arange(1,11)])
plt.plot(x,narr,'x')
x=np.arange(1,11)
narr=np.array([f(k, 0.5) for k in np.arange(1,11)])
plt.plot(x,narr,'o')
x=np.arange(1,11)
narr=np.array([f(k, 0.7) for k in np.arange(1,11)])
plt.plot(x,narr,'^')
x=np.arange(1,11)
narr=np.array([f(k, 0.9) for k in np.arange(1,11)])
plt.plot(x,narr,'*')
```

- #### ***기하분포 이름의 유래에 대해 간단히 알아보자***
  
  
      기하분포라는 이름은 기하수열에서 온 말이라고 한다. 기하 수열은 등비수열과 동치이며 기하수열(등비수열)은 또 기하 평균에서 시작한다고 한다. 모두 기하라는 말이 들어있는데 기하란 geometric 선이나 곡선 즉, 도형에 관한 것이다. 따라서 기하평균은 도형에서 발견한 평균이다. 이에 대해 간단히 알아보자


      
      두 변의 길이 a,b의 평균을 넓이의 관점에서 구하는 것이다.
      
      이때 b가 기하평균이다.
      
      기하수열은 수열, 수의 나열, 기하 수열의 특징은 중항의 제곱이 양 항의 곱으로 이루어지는데 아마 여기서 시작된 것 같다.
      
      기하 분포의 함수는 p(x)=(1-p)^(x-1)*p 인데 모양이 등비수열의 일반항 공식과 매우 흡사하다. 그래서 이 분포에 기하 분포가 붙었다고 한다.
