- 지수 분포를 유도해보자
- 길냥이 예제를 통해 지수 분포의 유도과정을 이해해보고 일반화 시켜보자
  - 하루 동안 길냥이를 만날 평균 횟수를 3회라고 해보자. 그렇다면 $\lambda$가 3이라는 것이고 이때 푸아송 분포는 아래와 같이 작성할 수 있겠다.

    - 길냥이를 마주칠때까지 걸리는 기간이 5일 이하일 확률을 구해보자
    - 
    - 길냥이를 만나는 기간이 5일이하면 되니까 1일차에 길냥이를 만나도 되고 1일차에 만나지 않고 2일차에 만나도 되고 또는 1,2일차 만나지 않고 3일차에 만나도 되고 또는 1,2,3일차에 만나지 않고 4일차에 만나도 되고, 마지막은 1,2,3,4에 만나지 않고 5일차에 만나도 된다. 이 다섯가지 확률을 다 더하면 왼쪽 확률을 구할 수 있다.
    - 두 번째 방법은 여사건을 이용하는 방법이다. 1에서 5일동안 길냥이를 마주치지 않을 확률을 빼주면 된다. 그러면 5일안에 길냥이를 마주칠 확률이 되는 것이다. 
      - 이 두 번째 방법을 이용해보자
      - 하루동안 길냥이를 마주치지 않을 확률 : 푸아송 분포 x에 0을 대입해주면 된다.
      - 이 하루동안 길냥이를 마주치지 않을 확률을 다섯 번 곱하면 5일동안 길냥이를 마주치지 않을 확률이 된다. 
      - 이 확률을 1에서 빼주면 두 번째 방법이 되는 것이다.
    - 일반화 해보자
      - 길냥이를 마주친다는 것은 어떤 사건이 발생한다는 것이기 때문에 여기에 어떤 사건이 발생할 때까지 걸리는 기간이 T일 이하일 확률
      - 
      - 이제 푸아송 분포를 일반화해보자. 
      - 평균발생횟수가 3회가 아니라 $\lambda$라고 했을 때,

      - 두 번째 방법을 이용한 것이다. 
      - 우리가 최종적으로 구해야하는 것은 f(t)이다.

- 지수분포의 평균을 구해보자
  - 먼저 지수분포의 함수식을 써보자
  - 
  - 지수분포에서는 확률변수가 시간이기 때문에 x대신에 t를 사용한다. 그리고 시간이기 때문에 0부터 무한대까지의 값을 갖는다. 
  - 
  - $ = \frac{1}{\lambda}$ (지수함수가 무한대로 가는 속도가 다항함수보다 빠르다는 것을 염두에 두자. 혹은 로피탈정리 또는 비교판정법을 이용하여 알 수도 있다.)

- 지수분포의 분산을 구해보자
  - 지수분포의 확률변수를 t라고 놓자.
  - 
  - 적분 구간이 0부터 무한대인 이유는 지수 분포의 확률 변수가 t로서 시간이기 때문에 그렇다.
  - 
  - 
  - 따라서 평균은 $\frac{1}{\lambda}$, 분산은 $\frac{1}{\lambda^2}$

- 지수분포의 누적분포함수

- 지수분포는 주로 누적분포함수의 형태로 사용된다. 누적분포함수는 영어로 Cumulative distribution function 즉 CDF라고 부른다. 
- 누적분포함수는 확률밀도함수를 적분하여 구하는데, 지수분포는 그럴 필요가 없다. 지수분포의 누적분포 함수를 이미 유도했기 때문이다. 우리는 지수분포를 유도하는 과정에서 지수분포의 누적 분포 함수를 먼저 유도하고 누적 분포함수에서 확률밀도함수를 끌어내는 방식을 사용했다. (아마 위에서 T로 양변을 미분하는 과정에서 integral안에 있던 f(t)를 끌어내는 과정을 말하는 것 같다)
