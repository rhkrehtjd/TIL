# 이산 확률 분포 7가지 총정리

- 확률 분포는 이산 확률 분포와 연속 확률 분포가 있다. 
  - 이산 확률 분포 7가지에 대해 배워봤다
      - 1) 베르누이 분포 : 시행횟수가 1번이며 이 시행횟수의 결과가 성공 혹은 실패로 나뉘는 분포이다.
      - 2) 이항 분포 : 베르누이 시행을 n번 할 때, 사건이 발생한 횟수를 확률변수 X로 하는 분포를 이항분포라 한다. 사건이 발생한 횟수는 성공 횟수라 할 수 있겠다.
      - 3) 기하 분포 : 성공이 나오기까지의 시행횟수를 확률변수 X로 하는 확률 분포이다. 그리고 각각의 시행은 베르누이 시행이다. 
      - 4) 음이항 분포 : 실패가 r번 나올 때까지 성공한 횟수가 x회인 분포
      - 5) 푸아송 분포 : 어떤 단위 시간동안 발생한 사건 횟수를 x라고 했을 때의 확률 분포, 이항 분포에서 n이 무한대, p가 0으로, np=$\lambda$일 때의 확률 분포이다. 
      - 6) 초기하 분포 : 모집단의 크기가 m이고 우리가 원하는 원소는 k개 들어있다. 이 모집단에서 크기가 n인 표본을 뽑았을 때, 이 표본 안에 들어있는 우리가 원하는 원소의 개수를 확률 변수 x라고 했을 때의 확률 분포이다.
      - 7) 다항 분포 : 이항 분포의 확장 버전이다. 어떤 시행의 결과가 3개 이상인 경우를 말한다.
