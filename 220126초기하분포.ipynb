# 초기하 분포

- 모집단의 원소의 개수가 m인 즉 크기가 m인 모집단이 있을 때, 이 모집단 안에는 우리가 원하는 원소가 k개가 들어있다. 이 모집단에서 크기가 n인 표본을 뽑았을 때 이 표본안에 우리가 원하는 원소가 x개 들어있을 확률 분포가 초기하 분포이다.
- 초기하 분포를 P(X)로 놓는다면 P(0)은 이 모집단에서 표본을 n개 뽑았는데 우리가 원하는 원소가 1개도 없을 확률이다. P(1)은 우리가 원하는 원소가 1개 들어있는 확률.
- 실제 예를 들어보자
  - 대표적으로 로또가 있다. 이 로또라는 모집단 안에는 45개의 공이 있다. 즉 모집단의 크기가 45이다. 이 45개의 공 중에서 우리는 6개의 공을 원한다. 즉 여기서 k=6이다. 여기서 크기가 6인 표본을 뽑았고 우리가 써낸 6개 숫자중에서 이 표본안에 x개가 들어있을 확률분포가 초기하분포이다. P(0)은 6개의 공을 뽑았는데 우리가 써낸 공이 한 개도 없을 확률이다.

- 초기하분포의 유도
  - 크기가 m인 모집단이 있으며 이 모집단 안에는 우리가 원하는 원소가 k개 있다. 이 모집단에서 크기가 n인 표본을 추출. 모집단에서 표본을 추출하는 경우의 수를 구해보자. 조합을 이용하여 m COMBINATION n이다. 그리고 이 크기 n인 표본안에 우리가 원하는 원소가 x개가 들어있어야 한다. 이러한 경우의 수를 구해보자.
       - 이 x개는 k개에서 뽑힌 것이다.  이때 x개는 k개에서 뽑힌 것. 왜? 모집단에는 우리가 원하는 원소가 k개 있기 때문에. 그래서 먼저 이 k개에서 x개 뽑는다. 그런데 표본에는 우리가 원하는 원소만 있는 것이 아니다. 이때 우리가 원하지 않는 원소의 개수는 n-x개이다. n-x개는 어디서 뽑혔을까? 이 모집단에서 k개가 아닌 나머지 부분에서 뽑혔을 것이다. 즉 m-k개에서 뽑혔을 것이다. 즉, 표본의 원소 n개 중에서 x개는 모집단의 k개로부터 뽑힌 것이고 n-x개는 모집단의 m-x개에서 뽑혔을 것이다. 이 경우의 수를 풀어써보자.
       - 원하는 원소가 k개 들어있고, 크기가 m인 모집단에서 크기가 n인 표본을 뽑을 때 원하는 원소가 x개 들어있을 경우의 수
       - 
       - 그렇다면 이런 상황에서 확률은 어떻게 될까?
       - 특정 사건이 발생할 사건의 경우의 수/ 전체 경우의 수
       - 
       - 이식에서 x는 변수이고 m,k,n은 사전에 정해지는 상수이다.

- 초기하분포라는 이름이 붙은 이유에 대해 알아보자
  - 초기하분포는 초기하함수로부터 이름이 유래되었고, 초기하 함수는 기하 급수로부터 유래되었다. 기하 급수는 기하 수열의 합이다.

       - 초기하 함수라는 것은 여러 종류의 다양한 함수들을 포함하는 함수이다. 초기하 함수에 있는 어떤 변수를 특정하거나 극한으로 보내면 특수한 함수들이 되는 그런 일반적인 형태의 함수가 초기하 함수이다.
       - 비유를 하자면 우리가 타원이나 원이나 쌍곡선이나 포물선과 같은 이러한 이차 곡선들을 전부 포함하는 이차 곡선을 아래와 같다고 하자
       - 
       - 이러한 일반적인 형태에서 A와B를 같게 두면 원이 되고 A=0,B!=0일 땐 포물선이 되듯 A,B,C,D라는 상수에 조건을 주면 여러가지 곡선을 표현할 수 있게되듯 이러한 개념이 초기하 함수이다.

- 초기하분포와 이항분포의 차이
  - 간단한 예시를 통해 초기하 분포를 다른 관점에서 이해해보자
       - 어떤 상자안에 흰구슬이 3개 검정 구슬이 2개 들어있다고 하자. 이 상자안에서 3개의 공을 꺼낼건데 3개의 공을 꺼냈을 때 검정 구슬이 1개 포함되어 있을 확률을 구해보자. 
       - 이 확률은 초기하 분포이다.
       - 
       - 이번에는 같은 상자에서 공을 꺼내는데 공을 1개씩 3번 꺼내보자
       - 마찬가지로 검정색 공이 1번 나올 확률을 구할 것이며 비복원 추출이다.
       - 이 경우는 공이 뽑히는 순서에 따라 세가지 경우로 나뉠 것이다. 
       - 검흰흰, 흰검흰, 흰흰검의 경우가 있을 것이다. 
       - 각각의 경우의 확률을 구해보자
       - 
       - 즉, 두 경우가 동일함을 알 수 있다.
       - 따라서 우리는 다음과 같이 이해할 수 있겠다.
       - 크기가 m이고 우리가 원하는 원소가 k개 들어있는 모집단에서 크기가 1인 표본을 비복원 추출로 n번 뽑을 때 우리가 원하는 원소가 x개 들어있을 확률 분포라고도 초기하 분포를 설명할 수 있겠다. 
       - 이번에는 비복원 추출이 아니라 복원 추출로 공을 뽑아보자
       - 똑같은 모집단에서 크기가 1인 표본을 3번 뽑을 것이다.
       - 이때 검정색 공이 1개 나올 확률을 복원 추출로 구할 것이다.
       - 
       - 이항분포의 모양이다.
       - 그래서 초기하분포와 이항분포의 차이?
       - 크기가 m이고 우리가 원하는 원소가 k개 들어있는 모집단에서 크기가 1인 표본을 비복원 추출로 n번 뽑을 때 우리가 원하는 원소가 x개 들어있을 확률분포는 초기하 분포가 되고, 이때 비복원 추출을 복원추출로 변경하면 초기하 분포가 아니라 이항분포가 되는 것이다.
