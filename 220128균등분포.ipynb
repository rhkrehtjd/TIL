- 연속 확률 분포
  - (연속) 균등 분포 (Uniform Distribution) : 이산확률분포에서도 균등분포를 정의할 수 있기 때문에 앞에 (연속)이라고 이름을 붙여주었다.
  - 모든 확률 변수의 함수값이 동일한 분포이다. 
  - 확률 변수가 x_1부터 x_n까지 있다고 할 때 이 모든 함수값이 전부 동일한 분포이다. 
  - 이산 확률 분포에서는 확률분포함수값이 곧 확률이었지만 연속확률분포에서 확률분포함수값은 확률밀도라고 부르는 값이 된다.

- 균등분포의 예시
  - 5분마다 한대씩오는 버스 한 대가 있다. 우리가 집에 있다가 임의의 시간에 버스에 타기 위해 정류장에 나갔을 때 버스를 기다릴 시간이 x분일 확률을 구해보자
  - 버스를 기다리는 시간은 최소 0분에서 최대 5분일 것이다. 
  - 그런데 우리는 임의의 시간에 나가는 것이기 때문에 이 모든 확률은 동일하게 k일 것이다.
  - 이때 확률변수 x는 셀 수가 없는 확률변수이다. 셀 수 없다는 의미는 단순히 그 개수가 무한대라는 것과는 별개의 의미이다. 자연수도 무한개이다. 그런데 자연수는 셀 수 있다. 이때 확률변수 x에서 0다음에 몇분이 올지는 알 수 없다. 이것이 바로 셀 수 없다는 의미이다. 이렇게 셀 수 없는 확률 변수를 연속확률변수라고 한다. 연속확률변수에서는 함수값이 확률이 아니라 확률밀도값이된다. 그래서 우리는 이 범위에서의 확률밀도함수값을 k라 놓고 그래프를 한 번 그려보자
  - 
  - 확률밀도함수
  - 
  - 확률밀도함수이기 때문에 특정 시간 x의 확률을 정의할 순 없다. 
  - 예를 들어 버스를 기다릴 시간이 3분일 확률은 정의할 수 없다. 그냥 0이다. 존재하지 않는 것이다. 
  - 확률밀도함수에서 구할 수 있는 것은 어떤 범위에서의 확률값이다. 예를 들어 버스를 기다릴 시간이 0분에서 2분 사이일 확률은 $2\times\frac{1}{5} = \frac{2}{5}$임을 알 수 있다.
